#### program to print the maximum and minimum number

```python
numbers = [12,45,67,23,89,34]

largest = max(numbers)

print("list",numbers)
print("largest element:",largest)
```

```python
numbers = [12,45,67,23,89,34]

smallest = min(numbers)

print("list",numbers)
print("smallest element:",smallest)
```

#### program to print number of a character in the string

```python
text = "python"
frequency = {}
for ch in text:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] =1
print(frequency)
```

#### program to delete the duplicates

```python
numbers = [10,20,30,40,10,50]

unique_numbers = list(set(numbers))

print("original list",numbers)
print("unique list:",unique_numbers)
```

#### to reverse a string

```python
text = "engneering"

reversed_text = text[::-1]

print("original string:",text)
print("reversed string:",reversed_text)
```
